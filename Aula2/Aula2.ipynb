# Automação de tarefas
- Buscar dados de valores das ações
- Criar as análises solicitadas pelo gestor
- Enviar um e-mail com os resultados das análises

# Passo 1 - Buscar dados

```python
!pip install yfinance
```

```python
import yfinance
```

```python
ticker = input("Digite o código da ação: ").upper()
dados = yfinance.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

# Passo 2 - Criar a análise
- Cotação máxima
- Cotação mínima
- Cotação atual

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]
```

# Passo 3 - envio de e-mail
- Abrir uma nova aba e digitar o endereço do gmail + Enter
- Clicar no botão "Escrever"
- Preencher o e-mail 
- Preencher o assunto
- Escrever o email com os resultados da análise
- Clicar em "Enviar"

```python
!pip install pyautogui
!pip install pyperclip
```

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 2.5

#entrando no gmail
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

#colocar as informações do e-mail
pyautogui.click(x=106, y=166)
pyperclip.copy("gestor@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#escrever a mensagem
mensagem = f"""Bom dia!

Segue as análises dos últimos 6 meses da ação {ticker}

Cotação máxima: R$ {round(maxima, 2)}
Cotação mínima: R$ {round(minima, 2)}
Cotação atual: R$ {round(atual, 2)}"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")
pyautogui.click(x=763, y=743)
```
