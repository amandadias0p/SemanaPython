# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

### Entrar no sistema da empresa, fazer login e baixar o arquivo

```python
# já havia instalado o pyautogui em outra oportunidade então precisava somente importar
import pyautogui
import time
```

```python
pyautogui.PAUSE = 1

# entrar no sistema da empresa
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")
time.sleep(3)

# escrever login e senha
pyautogui.click(x=1017, y=344)
pyautogui.write("meu-login")
pyautogui.click(x=1014, y=419)
pyautogui.write("minha-senha")
pyautogui.click(x=1033, y=489)
time.sleep(2)

# baixar o arquivo
pyautogui.click(x=1311, y=288)
time.sleep(2)
pyautogui.click(x=1073, y=662)
time.sleep(2)
```

```python
time.sleep(5)
pyautogui.position()
```

### Calcular os indicadores

```python
import pandas as pd

tabela = pd.read_csv(r'C:/Users/AMANDA/Downloads/Compras.csv', sep=";") #para acessar a tabela, precisei dizer que o separador era ;
tabela["ValorFinal"] = tabela["ValorFinal"].str.replace(",", ".").astype(float) #transformando vírgulas em pontos
display(tabela)
```

```python
total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade
```

### Enviar o e-mail

```python
import pyperclip
# entrar no email
pyautogui.hotkey("ctrl", "t")
pyautogui.write("gmail.com")
pyautogui.press("enter")
time.sleep(5)

# clicar no botao escrever
pyautogui.click(x=791, y=181)

# preencher as informações do email
pyautogui.write("seu-email@gmail.com")
time.sleep(1)
pyautogui.press("tab")
pyautogui.press("tab")
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")

texto = f"""Prezados,
Segue o relatório de compras:
Total gasto: R$ {total_gasto:,.2f}
Quantidade: {quantidade}
Valor médio: R$ {preco_medio:,.2f}"""


pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# enviar
pyautogui.hotkey("ctrl", "enter")
```
